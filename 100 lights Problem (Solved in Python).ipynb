```python
def colored(r, g, b, text):
    return "\033[38;2;{};{};{}m{} \033[38;2;255;255;255m".format(r, g, b, text)

OFF = colored(255,0,0,"OFF")
ON = colored(0,255,0,"ON")

#Print initial state of all 100 lights
n = 100
lights = [False] * n
for i in range(1, n+1):
    print("Light#", i, "\t", "ON" if(lights[i-1]) else "OFF" ) 
```

```python
for i in range(1, n+1):
    print("\nFlip Light#", i, "and it's multiple") 
    for j in range(1, n+1):
        if(not j%i):
            print("Flip Light#", j) 
            state = lights[j-1]
            lights[j-1] = not state

print("\n")
for i in range(1, n+1):
    print("Final state of Light#", i, "\t", "ON" if(lights[i-1]) else "") 
    
```
